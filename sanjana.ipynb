```python
import matplotlib.pyplot as plt

# Values
states = ['West Bengal', 'Odisha', 'Maharashtra']
population_values = [valuer1, valuer2, valuer3]
percentage_values = [rper1, rper2, rper3]

# Create figure and axes
fig, ax1 = plt.subplots()

# Plot the population values
ax1.bar(states, population_values, color='b', alpha=0.6, label='Population')
ax1.set_ylabel('Population', color='b')
ax1.tick_params('y', colors='b')

# Create second y-axis for percentage values
ax2 = ax1.twinx()
ax2.plot(states, percentage_values, color='r', marker='o', label='Percentage')
ax2.set_ylabel('Percentage (%)', color='r')
ax2.tick_params('y', colors='r')

# Show the legend
fig.legend(loc="upper right")

# Show the plot
plt.title('Comparison of Rural Population')
plt.show()
```
