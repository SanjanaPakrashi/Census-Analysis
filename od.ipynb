```python
#importing sheets
import pandas as pd

statedf = pd.read_excel('state2.xlsx', header=[0, 1, 2,3,4])
state = "Odisha"
```

```python
#Prints Coloumn Names for Odisha
for col in statedf.columns:
    print(col)
```

```python
#Coloumn Names
age_group_column = ('Age-Group', 'Unnamed: 5_level_1', 'Unnamed: 5_level_2', 'Unnamed: 5_level_3', 'Unnamed: 5_level_4')
area_column = ('Area', 'Unnamed: 3_level_1', 'Name', 'Unnamed: 3_level_3', 'Unnamed: 3_level_4')
total_column = ('Total/', 'Unnamed: 4_level_1', 'Rural/', 'Urban', 'Unnamed: 4_level_4')
area_type=('Total/', 'Unnamed: 4_level_1', 'Rural/', 'Urban', 'Unnamed: 4_level_4')

population_column = ('Population', 'Unnamed: 6_level_1', 'Unnamed: 6_level_2', 'Unnamed: 6_level_3', 'Persons')
populationm_column = ('Population', 'Unnamed: 7_level_1', 'Unnamed: 7_level_2', 'Unnamed: 7_level_3', 'Males')
populationf_column = ('Population', 'Unnamed: 8_level_1', 'Unnamed: 8_level_2', 'Unnamed: 8_level_3', 'Females')

workers_column = ('Main workers', 'Unnamed: 9_level_1', 'Unnamed: 9_level_2', 'Unnamed: 9_level_3', 'Persons')
workersm_column = ('Main workers', 'Unnamed: 10_level_1', 'Unnamed: 10_level_2', 'Unnamed: 10_level_3', 'Males')
workersf_column = ('Main workers', 'Unnamed: 11_level_1', 'Unnamed: 11_level_2', 'Unnamed: 11_level_3', 'Females')

nonworkers_column = ('Non-workers', 'Unnamed: 21_level_1', '__________________________________________________ ', 'Total', 'Persons')

unemployednw_column = ('Non-workers', 'Unnamed: 24_level_1', '__________________________________________________ ', 'Seeking/availabe for work', 'Persons')
unemployedmw_column = ('Marginal workers', 'Worked for 3 to 6 months', 'Unnamed: 18_level_2', 'Seeking/available for work', 'Persons')
```

```python
sp = statedf[population_column].iloc[0]
print(f"Total Population in {state}: {sp}")
```

```python
# Population in Rural Area
age_group_condition = "Total"
area_condition = "State - ODISHA (21)"
total_condition = "Rural"

filteredr_df = statedf[
    (statedf[age_group_column] == age_group_condition) &
    (statedf[area_column] == area_condition) &
    (statedf[total_column] == total_condition)
]

valuer = filteredr_df[population_column].values[0]
rper=(valuer/sp) * 100

print(f"Rural Population in {state} : {valuer}")
print(f"Rural Population Percentage : {rper}")
```

```python
# Population in Urban Area
age_group_condition = "Total"
area_condition = "State - ODISHA (21)"
total_condition = "Urban"

filteredu_df = statedf[
    (statedf[age_group_column] == age_group_condition) &
    (statedf[area_column] == area_condition) &
    (statedf[total_column] == total_condition)
]

valueu = filteredu_df[population_column].values[0]
uper=(valueu/sp) * 100

print(f"Urban Population in {state} : {valueu}")
print(f"Urban Population Percentage : {uper}")
```

```python
import matplotlib.pyplot as plt

# Population percentages
pop_percentages = [rper, uper]

# Corresponding labels
labels = ['Rural', 'Urban']

# Specify colors for the pie
colors = ['#ff9999','#66b3ff']

# Create pie chart

plt.pie(pop_percentages, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, pctdistance=0.85, shadow=True)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

# Add a title
plt.title(f'Population Distribution in {state}')

# Display the chart
plt.show()
```

```python
total_workers =statedf[workers_column].iloc[0]
print(f"Total Working Population in {state}: {total_workers}")

wper=(total_workers/sp)*100
print(f"Total Working Population percentage in {state}: {wper} %")
```

```python
total_unemployed = statedf[unemployednw_column].iloc[0] + statedf[unemployedmw_column].iloc[0]
print(f"Unemployed seeking work in {state}: {total_unemployed}")

unper=(total_unemployed/sp)*100
print(f"Unemployment rate in {state}: {unper} %")
```

```python
labels = ['Employed', 'Unemployed']
sizes = [wper, unper]
colors = ['skyblue', 'yellowgreen']

# Plotting the bar graph
bars = plt.bar(labels, sizes, color=colors)
plt.xlabel('Employment Status')
plt.ylabel('Population Percentage(%)')
plt.title(f'Employed vs Unemployed Population in {state}')

# Adding the values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
#Total Male Working Population
swm = statedf[workersm_column].iloc[0]
print(f"Total Working Male Population in {state}: {swm}")

mwper=(swm/sp) * 100
print(f"Percentage of working population of Males in {state}: {mwper} %")
```

```python
#Total Female Working Population
swf = statedf[workersf_column].iloc[0]
print(f"Total Working Female Population in {state}: {swf}")

fwper=(swf/sp) * 100
print(f"Percentage of working population of Females in {state}: {fwper} %")
```

```python

# Assuming you have the data for male and female working population percentages
male_percentage = mwper
female_percentage = fwper

# Data to plot
labels = ['Males', 'Females']
percentages = [male_percentage, female_percentage]

# Create the bar graph
plt.bar(labels, percentages, color=['blue', 'pink'])

# Add title and labels
plt.title(f'Comparative Analysis of Male and Female Workforce Participation in {state} ')
plt.xlabel('Gender')
plt.ylabel('Percentage')

# Show the graph
plt.show()
```

```python
# Population of Senior Citizens
age_group_condition = "60+"
area_condition = "State - ODISHA (21)"
total_condition = "Total"

filtered_df = statedf[
    (statedf[age_group_column] == age_group_condition) &
    (statedf[area_column] == area_condition) &
    (statedf[total_column] == total_condition)
]

value = filtered_df[population_column].values[0]
snrper=(value/sp) * 100

print(f"Senior Citizens in {state} : {value}")
print(f"Senior Citizen Percentage : {snrper}")
```

```python
age_group = statedf[age_group_column]
population = statedf[population_column]

# Filter the DataFrame based on 'Area' and 'Total' columns
filtered_df = statedf[(statedf[area_column] == 'State - ODISHA (21)') & (statedf[total_column] == 'Total')]

# Exclude the age groups 'Total', '15-59', '60+'
excluded_groups = ['Total', '15-59', '60+']
filtered_df = filtered_df[~filtered_df[age_group_column].isin(excluded_groups)]

# Get the 'Age-Group' and 'Population' data from the filtered DataFrame
age_group = filtered_df[age_group_column]
population = filtered_df[population_column]

# Calculating the percentage of population for each age group
percentage_population = (population / sp) * 100

# Creating a new DataFrame for plotting
plot_data = pd.DataFrame({
    'Age-Group': age_group,
    'Population': population,
    'Percentage of Population': percentage_population
})

# Printing the values
print(plot_data)

# Plotting the data
plt.figure(figsize=(10, 6))
bars = plt.bar(plot_data['Age-Group'], plot_data['Percentage of Population'], color='skyblue')

# Adding the values to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 2), ha='center', va='bottom')

plt.xlabel('Age Group')
plt.ylabel('Percentage of Population (%)')
plt.title('Percentage of Population by Age Group in West Bengal (Excluding Total, 15-59, 60+)')
plt.xticks(rotation=45)
plt.show()
```

```python
newfilter = statedf.drop_duplicates(subset=[area_column], keep='first')
population_values = newfilter[population_column].tolist()
population_values = population_values[1:]
print(population_values)
```

```python

modified_population_values = [(value / sp) * 100 for value in population_values]
print(modified_population_values)
district_values = newfilter[area_column].tolist()
district_values = district_values[1:]
print(district_values)

district_values = [v[11:-5] for v in district_values]

print(district_values)
```

```python

districts = district_values
pops = modified_population_values

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(districts, pops, color='blue')

# Adding titles and labels
plt.title('Population of Districts')
plt.xlabel('Districts')
plt.ylabel('Population')

plt.xticks(rotation=90, ha='right')

# Showing the plot
plt.tight_layout() 
plt.show()
```
